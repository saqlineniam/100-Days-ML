```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'train.csv')
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
df.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace = True)
```

```python
df.head(5)
```

```python
df.info()
```

```python
si = SimpleImputer()
si.fit(df[['Age']])
```

```python
df['Age'] = si.transform(df[['Age']]).round(1)
```

```python
df.head()
```

```python
df.info()
```

```python
si_embarked = SimpleImputer(strategy = 'most_frequent')
```

```python
si_embarked.fit(df[['Embarked']])
```

```python
df['Embarked'] = si_embarked.transform(df[['Embarked']]).ravel()
```

```python
df.info()
```

```python
df['Pclass'].value_counts()
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe_sex = OneHotEncoder(handle_unknown = 'ignore', sparse_output = False)
ohe_embarked = OneHotEncoder(handle_unknown = 'ignore', sparse_output = False)
```

```python
ohe_sex.fit(df[['Sex']])
```

```python
ohe_embarked.fit(df[['Embarked']])
```

```python
sex = ohe_sex.transform(df[['Sex']])
embarked = ohe_embarked.transform(df[['Embarked']])
```

```python
sex = pd.DataFrame(sex)
```

```python
embarked = pd.DataFrame(embarked)
```

```python
df.drop(columns = ['Sex', 'Embarked'], inplace = True)
```

```python
df = pd.concat([df,sex,embarked], axis = 1)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns = ['Survived']), df['Survived'], test_size = 0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
sex_feature_names = [f'sex_{cat.lower()}' for cat in ohe_sex.categories_[0]]
embarked_feature_names = [f'embarked_{cat.lower()}' for cat in ohe_embarked.categories_[0]]

# Assign column names to the encoded DataFrames
sex.columns = sex_feature_names
embarked.columns = embarked_feature_names

# Now concatenate with proper column names
df = pd.concat([df, sex, embarked], axis=1)

# Convert all column names to strings
df.columns = df.columns.astype(str)

# Now split and fit should work
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), 
                                                    df['Survived'], 
                                                    test_size=0.2, 
                                                    random_state=42)

clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```
