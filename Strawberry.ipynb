```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.graphics.gofplots import qqplot
from statsmodels.graphics.factorplots import interaction_plot
from scipy.stats import levene
```

```python
df = pd.read_excel(r"C:\Users\saqli\Desktop\FULL DataSet.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:, 0:4]
y = df.iloc[:, 4:14]
```

```python
from sklearn.preprocessing import OrdinalEncoder
Oe = OrdinalEncoder()
```

```python
Oe.fit(x[['Days']])
```

```python
x['Days'] = Oe.transform(x[['Days']])
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
y = y[['Firmness']]
```

```python
ss = StandardScaler()
ss.fit(y[['Firmness']])
y = ss.transform(y[['Firmness']])
```

```python
y = pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_test.shape
```

```python
svr = SVR(kernel='rbf', C=300, epsilon=0.1)
svr.fit(x_train, y_train)

# 5. Prediction
y_pred_scaled = svr.predict(x_test)

# 6. Evaluation
print("MSE:", mean_squared_error(y_test, y_pred_scaled))
print("R2 Score:", r2_score(y_test, y_pred_scaled))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
print("MSE:", mean_squared_error(y_test, y_pred_lr))
print("R2 Score:", r2_score(y_test, y_pred_lr))
print('MAE', np.mean(np.abs(y_test - y_pred_lr)))
print('RMSE', np.sqrt(mean_squared_error(y_test, y_pred_lr)))
```
