```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'placement (1).csv')
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(x_train, lr.predict(x_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred = lr.predict(x_test)
```

```python
print('MAE: ', mean_absolute_error(y_test, y_pred))
```

```python
print('MSE', mean_squared_error(y_test, y_pred))
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
print('r2_score: ', r2_score(y_test, y_pred))
```

```python
r2_score = r2_score(y_test, y_pred)
```

```python
print('Adjusted R2 score: ', 1 - (1 - r2_score) * (len(y_test) - 1) / (len(y_test) - x_test.shape[1] - 1))
```

ADDING A RANDOM COLUMN

```python
new_df = df.copy()
```

```python
new_df['random_feature'] = np.random.random(200)
```

```python
new_df = new_df[['cgpa', 'random_feature', 'package']]
```

```python
new_df.head()
```

```python
plt.scatter(new_df['random_feature'], new_df['package'])
plt.xlabel('random_feature')
plt.ylabel('Package')
```

```python
x = new_df.iloc[:,0:2]
y = new_df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('r2_score: ', r2_score(y_test, y_pred))
```

```python
r2_score = r2_score(y_test, y_pred)
```

```python
print('Adjusted R2 score: ', 1 - (1 - r2_score) * (len(y_test) - 1) / (len(y_test) - x_test.shape[1] - 1))
```
